```python
import pandas as pd
from pathlib import Path
import numpy as np
import os
from libs.p1_library import chemical_filter
```

```python
def chemical_filter(file_path,chemical_list):
    chemical_data = []

    for filename in os.listdir(file_path):
        if filename.endswith(".csv"):
            csv_data = pd.read_csv(filename, parse_dates=True, infer_datetime_format=True,delimiter=',')
            for item in chemical_list:
                chemicals_filtered = csv_data[csv_data['CHEMICAL_NAME'] == item]
                chemicals_filtered = chemicals_filtered.iloc[: , 1:]
                chemical_data.append(chemicals_filtered)

    chemicals_df = pd.concat(chemical_data)
    return chemicals_df
```

```python
chemical_list = ["2,3,7,8-Tetrachlorodibenzo-p-dioxin","Dieldrin","Hexachlorobiphenyl; 3,3',4,4',5,5'- (PCB 169)","Pentachlorobiphenyl; 3,3',4,4',5- (PCB 126)","Mercury","Lead","Cyanide"]
file_path = '../data/cleandata/'
chemical_filter(file_path,chemical_list)
```
